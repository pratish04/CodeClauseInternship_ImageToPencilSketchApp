```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
actual_image = cv2.imread('image.jpg')
cv2_imshow(actual_image)
```

To convert an image into sketch, five operations need to be performed.
1. Convert the image into grayscale.

```python
grey_image = cv2.cvtColor(actual_image, cv2.COLOR_BGR2GRAY)
cv2_imshow(grey_image)
```

2. Invert the grayscale image.

```python
invert = cv2.bitwise_not(grey_image)
cv2_imshow(invert)
```

3. Blur the image.

```python
blur = cv2.GaussianBlur(invert, (21, 21), 0)
cv2_imshow(blur)
```

4. Invert the blurred image.

```python
invertedblur = cv2.bitwise_not(blur)
cv2_imshow(invertedblur)
```

5. Perform per-element division of the grayscale image array and the inverted blur image array with a scale of 256. <br/>
Print the final result.

```python
sketch = cv2.divide(grey_image, invertedblur, scale=256.0)
cv2_imshow(sketch)
```
